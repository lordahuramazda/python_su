```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table
plt.ion()
import os
```

```python
t = Table.read ("data.fits")
ra = t["RA"]
dec = t["DEC"]
av = t["Av_p50"]
#t[0].colnames
```

```python
plt.figure()
#plt.plot(ra, dec, '.b',ls='')
cb =plt.scatter(ra, dec, c=av, marker='o', cmap='magma', vmin=0, vmax=1)
plt.colorbar(cb, label='A(V)') 
plt.xlabel("RA", fontsize=14)
plt.ylabel("DEC", fontsize=14)
plt.savefig("lmc_av_spatial.eps")
```

```python
dist=t["distance_p50"]
```

```python
plt.figure()
plt.hist(dist/1000, bins=25)
plt.xlabel("Distance [kpc]", fontsize=14)
plt.ylabel("N")
d_mean= np.mean(dist/1000)
print(d_mean)
plt.axvline(d_mean, color='k', ls='--', lw=1)
plt.text(58, 270, '%s' % np.around(d_mean, decimals=2))
plt.savefig("hist.pdf")
```

```python
cat = Table.read("scylla.fits")
f = "scylla.fits"
```

```python
#t[0].colnames
f475 = cat["F475W_VEGA"]
f814 = cat["F814W_VEGA"]
```

```python
col = f475 - f814
mag = f475
n = len(f475)
plt.figure()
plt.plot(col, mag, ',b',ls='', label='N = %s' % n)
plt.legend()
plt.xlim(-2, 5)
plt.ylim(30, 17)
plt.xlabel('F475W - F814W')
plt.ylabel('F475W')
plt.title('%s' % f)
plt.savefig("cmd_scylla.pdf")
```
